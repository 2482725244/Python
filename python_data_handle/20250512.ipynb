## 1.上海温度展示

## 2.上海温度展示2.0

```python
import matplotlib.pyplot as plt 
import random
```

```python
plt.rcParams["font.family"] = ["SimHei"]
x = range(60)
y_shanghai = [random.uniform(20,30) for i in x] 
```

```python
x_ticks = ["十一点{}分".format(i) for i in x] 
y_ticks = ["{}度".format(i) for i in range(20,30)]

plt.figure(figsize=(20,8),dpi=100)
#添加刻度ticks
plt.xticks(x[::5],x_ticks[::5])
plt.yticks(range(20,30),y_ticks)
#添加网格
plt.grid(True, linestyle='--',alpha = 1)
#添加标注信息
plt.ylabel("温度")
plt.xlabel("时间",fontsize = "20")
plt.title("上海十一点至十二点温度变化折线图",fontsize = "20")
plt.plot(x,y_shanghai)
plt.savefig("shanghai.png")
plt.show()
```

## 3.温度展示3.0

```python
import matplotlib.pyplot as plt 
import random
plt.rcParams["font.family"] = ["SimHei"]
```

```python
y_shanghai = [random.uniform(20,30) for i in range(60)]
x_shanghai = range(60)
y_beijing = [random.uniform(10,30) for i in range(60)]
x_beijing = range(60)
x_ticks = ["11点{}分".format(i) for i in range(60)]
plt.figure(figsize=(20,8),dpi=100)
plt.xticks(x_shanghai[::5],x_ticks[::5])
plt.yticks(range(10 ,31))
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("北京上海温度对比表")
plt.plot(x_shanghai,y_shanghai,color='b',linestyle='--',label = '上海')
plt.plot(x_beijing,y_beijing,color = 'r',linestyle='-', label = '北京')
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

## 4.多坐标系创建

```python
import matplotlib.pyplot as plt 
import random
plt.rcParams["font.family"] = ["SimHei"]
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi = 100)

y_shanghai = [random.uniform(20,30) for i in range(60)]
x_shanghai = range(60)
y_beijing = [random.uniform(10,30) for i in range(60)]
x_beijing = range(60)
x_ticks = ["11点{}分".format(i) for i in range(60)]
axes[0].plot(x_shanghai,y_shanghai,label="上海")
axes[1].plot(x_beijing,y_beijing,label="北京",color = 'r')

axes[0].set_xticks(x_shanghai[::5])
axes[0].set_yticks(range(20 ,31))
axes[0].set_xticklabels(x_ticks[::5])

axes[1].set_xticks(x_beijing[::5])
axes[1].set_yticks(range(10  ,31))
axes[1].set_xticklabels(x_ticks[::5])

axes[0].legend(loc=0)
axes[1].legend(loc=0)

axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[0].set_title("上海")

axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")
axes[1].set_title("北京")

axes[0].grid(True)
axes[1].grid(True)

plt.savefig("d.jpg")
plt.show()


# plt.figure(figsize=(20,8),dpi=100)
# plt.xticks(x_shanghai[::5],x_ticks[::5])
# plt.yticks(range(10 ,31))
# plt.xlabel("时间")
# plt.ylabel("温度")
# plt.title("北京上海温度对比表")
# plt.plot(x_shanghai,y_shanghai,color='b',linestyle='--',label = '上海')
# plt.plot(x_beijing,y_beijing,color = 'r',linestyle='-', label = '北京')
# plt.legend(loc="best")
# plt.grid(True)
# plt.show()
```

## 14.矩阵的运算

```python
#规则为，（m行，n列）*（n行，l列）=（m行，l列）
#否则都是相同结构的矩阵才能相乘
```

```python
aa = np.array([[34,28],[45,67],[99,98]])
aa
```

```python
bb = np.array([[0.6],[0.4]])
bb
```

```python
np.matmul(aa,bb)
```

```python
np.dot(aa,bb)
```

```python
v = 10
np.dot(aa,v)
```

## 1.上海温度展示

```python
import matplotlib.pyplot as plt
import random
```

```python
x = range(60)
y_shanghai = [random.uniform(20,30) for i in x]

plt.figure(figsize=(20,8),dpi=100)
plt.plot(x,y_shanghai)
plt.show()
```

## 2.上海温度展示2.0

```python
import matplotlib.pyplot as plt 
import random
```

```python
plt.rcParams["font.family"] = ["SimHei"]
x = range(60)
y_shanghai = [random.uniform(20,30) for i in x] 
```

```python
x_ticks = ["十一点{}分".format(i) for i in x] 
y_ticks = ["{}度".format(i) for i in range(20,30)]

plt.figure(figsize=(20,8),dpi=100)
#添加刻度ticks
plt.xticks(x[::5],x_ticks[::5])
plt.yticks(range(20,30),y_ticks)
#添加网格
plt.grid(True, linestyle='--',alpha = 1)
#添加标注信息
plt.ylabel("温度")
plt.xlabel("时间",fontsize = "20")
plt.title("上海十一点至十二点温度变化折线图",fontsize = "20")
plt.plot(x,y_shanghai)
plt.savefig("shanghai.png")
plt.show()
```
